```python
mass = pd.read_csv("physobjects/annotations/crowdsourced/mass/train.csv")
material = pd.read_csv("physobjects/annotations/crowdsourced/material/train.csv")



material_dict = {}
for row in material.index:
    material_dict.update({material["annotation_id_0"][row]:material["response"][row]})

prompts = []

for row in mass.index:
    response = mass["response"][row]
    if response == "unclear":
        continue
    elif response == "left":
        response_text = "is heavier"
    elif response == "right":
        response_text = "is lighter"
    elif response == "equal":
        response_text = "has the same mass"

    obj1 = category_dict[instance_category_dict[int(mass["annotation_id_0"][row])]]
    obj2 = category_dict[instance_category_dict[int(mass["annotation_id_1"][row])]]
    obj1_mat = material_dict.get(mass["annotation_id_0"][row])
    obj2_mat = material_dict.get(mass["annotation_id_1"][row])
    if obj1 == obj2:
        continue
    obj1_mat = "" if obj1_mat is None or obj1_mat == 'unknown' else obj1_mat
    obj2_mat = "" if obj2_mat is None or obj2_mat == 'unknown' else obj2_mat
    print(f"a {obj1_mat} {obj1} {response_text} than a {obj2_mat} {obj2}")
```
