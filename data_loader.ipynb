```python
mass = pd.read_csv("physobjects/annotations/crowdsourced/mass/train.csv")
material = pd.read_csv("physobjects/annotations/crowdsourced/material/train.csv")



material_dict = {}
for row in material.index:
    material_dict.update({material["annotation_id_0"][row]:material["response"][row]})

mass_prompts = []

for row in mass.index:
    response = mass["response"][row]
    if response == "unclear":
        continue
    elif response == "left":
        response_text = "is heavier"
    elif response == "right":
        response_text = "is lighter"
    elif response == "equal":
        response_text = "have the same mass"

    obj1 = category_dict[instance_category_dict[int(mass["annotation_id_0"][row])]]
    obj2 = category_dict[instance_category_dict[int(mass["annotation_id_1"][row])]]
    obj1_mat = material_dict.get(mass["annotation_id_0"][row])
    obj2_mat = material_dict.get(mass["annotation_id_1"][row])
    if obj1 == obj2:
        continue
    obj1_mat = "" if obj1_mat is None or obj1_mat == 'unknown' else obj1_mat
    obj2_mat = "" if obj2_mat is None or obj2_mat == 'unknown' else obj2_mat
    user_content = f"which has more mass: a {obj1_mat} {obj1} or a {obj2_mat} {obj2}?"
    if response == "equal":
        assistant_content = f"they {response_text}"
    else:
        assistant_content = f"the {obj1_mat} {obj1} {response_text}"
    # {"messages": [{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}]}
    mass_prompts.append([{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}])

```

```python
fragility = pd.read_csv("physobjects/annotations/crowdsourced/fragility/train.csv")
fragility_prompts = []

for row in fragility.index:
    response = fragility["response"][row]
    if response == "unclear":
        continue
    elif response == "left":
        response_text = "is more fragile"
    elif response == "right":
        response_text = "is less fragile"
    elif response == "equal":
        response_text = "have the same fragility"

    obj1 = category_dict[instance_category_dict[int(fragility["annotation_id_0"][row])]]
    obj2 = category_dict[instance_category_dict[int(fragility["annotation_id_1"][row])]]
    obj1_mat = material_dict.get(mass["annotation_id_0"][row])
    obj2_mat = material_dict.get(mass["annotation_id_1"][row])
    if obj1 == obj2:
        continue
    obj1_mat = "" if obj1_mat is None or obj1_mat == 'unknown' else obj1_mat
    obj2_mat = "" if obj2_mat is None or obj2_mat == 'unknown' else obj2_mat
    # print(f"a {obj1_mat} {obj1} {response_text} than a {obj2_mat} {obj2}")
    user_content = f"which is more fragile: a {obj1_mat} {obj1} or a {obj2_mat} {obj2}?"
    if response == "equal":
        assistant_content = f"they {response_text}"
    else:
        assistant_content = f"the {obj1_mat} {obj1} {response_text}"
    # {"messages": [{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}]}
    fragility_prompts.append([{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}])
```

```python
deformability = pd.read_csv("physobjects/annotations/crowdsourced/deformability/train.csv")
deformability_prompts = []

for row in deformability.index:
    response = deformability["response"][row]
    if response == "unclear":
        continue
    elif response == "left":
        response_text = "is more deformable"
    elif response == "right":
        response_text = "is less deformable"
    elif response == "equal":
        response_text = "have the same deformability"

    obj1 = category_dict[instance_category_dict[int(deformability["annotation_id_0"][row])]]
    obj2 = category_dict[instance_category_dict[int(deformability["annotation_id_1"][row])]]
    obj1_mat = material_dict.get(mass["annotation_id_0"][row])
    obj2_mat = material_dict.get(mass["annotation_id_1"][row])
    if obj1 == obj2:
        continue
    obj1_mat = "" if obj1_mat is None or obj1_mat == 'unknown' else obj1_mat
    obj2_mat = "" if obj2_mat is None or obj2_mat == 'unknown' else obj2_mat
    # print(f"a {obj1_mat} {obj1} {response_text} than a {obj2_mat} {obj2}")
    user_content = f"which is more deformable: a {obj1_mat} {obj1} or a {obj2_mat} {obj2}?"
    if response == "equal":
        assistant_content = f"they {response_text}"
    else:
        assistant_content = f"the {obj1_mat} {obj1} {response_text}"
    # {"messages": [{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}]}
    deformability_prompts.append([{"role": "system", "content": ""}, {"role": "user", "content": f"{user_content}"}, {"role": "assistant", "content": f"{assistant_content}"}])
```

```python
import random
seed = 42
random.seed(seed)

# shuffle the prompts
random.shuffle(mass_prompts)
random.shuffle(fragility_prompts)
random.shuffle(deformability_prompts)

# sample 250 samples from each prompt
n = 2500
mp = mass_prompts[:n]
fp = fragility_prompts[:n]
dp = deformability_prompts[:n]

# create a train and test set
ratio = 0.8
n_train = int(n * ratio)
train = mp[:n_train] + fp[:n_train] + dp[:n_train]
test = mp[n_train:] + fp[n_train:] + dp[n_train:]

# shuffle the train and test set
random.shuffle(train)
random.shuffle(test)
```

```python
train_sessions = train
test_sessions = test
with open('train-dg.jsonl', 'w') as f:
  for i, session in enumerate(train_sessions):
    print(session)
    json.dump({"messages": session}, f)
    if i < len(train_sessions) - 1:
      f.write('\n')
  print("Train dataset size:", len(train_sessions))

with open('test-dg.jsonl', 'w') as f:
  for i, session in enumerate(test_sessions):
    print(session)
    json.dump({"messages": session}, f)
    if i < len(test_sessions) - 1:
      f.write('\n')
  print("Test dataset size:", len(test_sessions))
```

```python
from openai import OpenAI
import time
openai.api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI()

train_file = client.files.create(file=open("train-dg.jsonl", "rb"), purpose="fine-tune")
test_file = client.files.create(file=open("test-dg.jsonl", "rb"), purpose="fine-tune")

# Start fine-tuning job.
ftjob = client.fine_tuning.jobs.create(training_file=train_file.id, validation_file=test_file.id, model="gpt-3.5-turbo")

# Track fine-tuning job until complete.
while ftjob.status != "succeeded":
  ftjob = client.fine_tuning.jobs.retrieve(ftjob.id)
  print(time.ctime(), "Status:", ftjob.status)
  time.sleep(30)
fine_tuned_model = ftjob.fine_tuned_model
```
